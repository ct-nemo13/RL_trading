```python
# -*- coding: utf-8 -*-
"""
Created on Sat Dec 30 16:48:20 2023

@author: ritwi
"""

import pandas as pd
import matplotlib.pyplot as plt


# Training the Default Agent
policies = ['default_LSTM_v2'] # 'default_LSTM_v1','default_LSTM_v2', 
training_range = 100

agents_per_policy = 3

result_parameters = ['rew_mean', 'rew_max', 'rew_min', 'vf_loss', 'policy_loss']


for result_parameter in result_parameters:
    plt.axhline(0, color='red', linestyle='--', label='Zero line')
    for policy in policies:
        for i in range(agents_per_policy):
            logs = pd.read_csv('training_logs/policy_' + policy + '/agent_' + str(i))

            # Plotting Column1 and Column2 with Time as the x-axis
            plt.plot(logs['agent_name'], logs[result_parameter], label=result_parameter+ '_' + policy + str(i))
            
            # Adding labels and legend
            plt.xlabel('agent_name')
            plt.ylabel(result_parameter)
            plt.legend()

    # Display the plot
    plt.show()
```
